```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("science.csv")

# Drop unnecessary column
df.drop(columns=['Unnamed: 0'], inplace=True)

# Handle missing values (optional, based on your need)
df['sex'] = df['sex'].fillna('Unknown')
```

```python
# 1. Bar Chart - Count of Public vs Private School
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='PrivPub', palette='pastel')
plt.title('Number of Students by School Type')
plt.xlabel('School Type')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# 2. Pie Chart - Gender Distribution
plt.figure(figsize=(6, 6))
gender_counts = df['sex'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("pastel"))
plt.title('Gender Distribution')
plt.tight_layout()
plt.show()
```

```python
# 3. Bar Chart - Average "like" Score by Gender
plt.figure(figsize=(6, 4))
sns.barplot(data=df, x='sex', y='like', palette='coolwarm')
plt.title('Average Interest in Science by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Like Score')
plt.tight_layout()
plt.show()
```

```python
# 4. Heatmap - Correlation Matrix
plt.figure(figsize=(6, 5))
sns.heatmap(df[['school', 'class', 'like', 'Class']].corr(), annot=True, cmap='YlGnBu', fmt=".2f")
plt.title('Correlation Between Numeric Features')
plt.tight_layout()
plt.show()
```
