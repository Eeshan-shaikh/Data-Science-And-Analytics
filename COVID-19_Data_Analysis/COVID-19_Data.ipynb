```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv("country_wise_latest.csv")
```

```python
# Sort by highest confirmed cases
top_countries = df.sort_values(by="Confirmed", ascending=False).head(10)
```

```python
# ----- Line Chart: Confirmed, Recovered, Deaths -----
plt.figure(figsize=(12, 6))
plt.plot(top_countries["Country/Region"], top_countries["Confirmed"], label='Confirmed', marker='o')
plt.plot(top_countries["Country/Region"], top_countries["Recovered"], label='Recovered', marker='s')
plt.plot(top_countries["Country/Region"], top_countries["Deaths"], label='Deaths', marker='x')
plt.xticks(rotation=45)
plt.ylabel("Cases")
plt.title("Top 10 Countries: Confirmed, Recovered, and Deaths")
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
# ----- Area Chart: Active Cases in Top 10 Countries -----
plt.figure(figsize=(12, 6))
plt.stackplot(top_countries["Country/Region"], 
              top_countries["Active"], 
              labels=['Active Cases'], colors=['#FFA07A'])
plt.title("Top 10 Countries: Active COVID-19 Cases")
plt.xlabel("Country")
plt.ylabel("Active Cases")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# ----- Heatmap: Rate Metrics -----
metrics = ['Deaths / 100 Cases', 'Recovered / 100 Cases', 'Deaths / 100 Recovered']
rate_df = top_countries.set_index('Country/Region')[metrics]

plt.figure(figsize=(10, 6))
sns.heatmap(rate_df, annot=True, fmt=".2f", cmap="YlGnBu")
plt.title("COVID-19 Outcome Rates per Country")
plt.tight_layout()
plt.show()
```
