```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv("student_performance_dataset.csv")

# ---- BASIC METRICS ----
print("Average Final Exam Score:", df['Final_Exam_Score'].mean())
print("Average Attendance Rate:", df['Attendance_Rate'].mean())
```

```python
# ---- CORRELATION ANALYSIS ----
correlation = df[['Final_Exam_Score', 'Attendance_Rate', 'Past_Exam_Scores', 'Study_Hours_per_Week']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
# ---- PASS vs FAIL COMPARISON ----
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Pass_Fail', hue='Pass_Fail', palette='Set2', legend=False)
plt.title("Pass vs Fail Distribution")
plt.show()
```

```python
# ---- ATTENDANCE IMPACT ----
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Pass_Fail', y='Attendance_Rate', hue='Pass_Fail', palette='pastel', legend=False)
plt.title("Attendance Impact on Performance")
plt.show()
```

```python
# ---- TOP vs STRUGGLING STUDENTS ----
top_students = df[df['Final_Exam_Score'] >= 70]
struggling_students = df[df['Final_Exam_Score'] < 50]
```

```python
# Bar chart: top vs struggling
labels = ['Top Students (>=70)', 'Struggling (<50)']
counts = [len(top_students), len(struggling_students)]

plt.figure(figsize=(8, 5))
sns.barplot(x=labels, y=counts, palette='husl')
plt.title("Top vs Struggling Students")
plt.ylabel("Number of Students")
plt.show()
```

```python
# ---- STUDY HOURS ANALYSIS ----
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Study_Hours_per_Week', y='Final_Exam_Score', hue='Pass_Fail', palette='cool')
plt.title("Study Hours vs Final Exam Score")
plt.grid(True)
plt.show()
```
